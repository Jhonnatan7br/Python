__Añadiendo funcionalidad al botón__

```python
#Preparación
class App(tk.Tk): 
    def __init__(self):
        super().__init__()

        # configurando la ventana
        self.title('Mi increíble App OMG') 
        self.geometry('500x500')
        self.resizable(False,False)
        
        #ENTRY
        self.cajita = ttk.Entry(self)
        self.cajita.pack()
        
        #BUTTÓN
        self.mibotonsito = ttk.Button(self, 
                                      text="Presióname", 
                                      command=self.mifuncion)
        self.mibotonsito.pack()
        
        #LABEL
        self.mi_variable_resultado=tk.StringVar() #Creo una variable control asociada al widget Label para guardar y actualizar la información
        self.label_resultado = ttk.Label(self, 
                                         font=("Helvetiva",40),
                                         textvariable=self.mi_variable_resultado) #Creo el Widget y asocio la variable
        self.label_resultado.pack()
    
    def mifuncion(self):
        #1. Guardar la información
        texto_usuario=self.cajita.get() #Get es por defecto el comando para "obtener" información
        #2. Hago el cálculo que quiero
        texto_invertido = texto_usuario[::-1] #Vamos a invertir el texto
        print(texto_invertido) #por verificar el punto 2
        #3. Muestro el resultado
        self.mi_variable_resultado.set(texto_invertido) #le damos un valor a la variable control del Label
        #4. Resetear parámetros (opcional)
        self.cajita.delete(0, END)
        
        

#Acción:
app = App()
app.mainloop()
```

# 8. MENU

```python
class App(tk.Tk): 
    def __init__(self):
        super().__init__()

        # configurando la ventana
        self.title('Mi increíble App OMG') 
        self.geometry('500x500')
        self.resizable(False,False)
        
        #Menu:
        self.createMenu()
        
    def createMenu(self):
        menuBar = Menu(self) #creamos un objeto de la clase Menu
        self.config(menu = menuBar) #configuramos ese nuevo objeto como el Menú de nuestra ventana
         
        #Creamos la primera categoría del Menú    
        archivo = Menu(menuBar, tearoff = 0) #CREAR CATEGORÍA
 
        menuBar.add_cascade(label="Archivo", menu=archivo) #nombre y dónde agregarla
        archivo.add_command(label='Nuevo')
        archivo.add_command(label="Abrir")
        archivo.add_separator()
        archivo.add_command(label='Salir', command=self.destroy)#para salir utilizar esa función
        
 
        # Creamos la segunda categoría
        help_menu = Menu(menuBar, tearoff = False) #Crear categoría
        
        menuBar.add_cascade(label = 'Help', menu = help_menu)
        help_menu.add_command(label = 'About')

        
#Acción:
app = App()
app.mainloop()
```

```python
class App(tk.Tk): 
    def __init__(self):
        super().__init__()

        # configurando la ventana
        self.title('Mi increíble App OMG') 
        self.geometry('500x500')
        self.resizable(False,False)
        
        #Menu:
        self.createMenu()
        
    def createMenu(self):
        mimenu = Menu(self) #creamos un objeto de la clase Menu
        self.config(menu = mimenu) #configuramos ese nuevo objeto como el Menú de nuestra ventana
         
        #Creamos la primera categoría del Menú    
        archivo_menu = Menu(mimenu, tearoff = 0) #CREAR CATEGORÍA
 
        mimenu.add_cascade(label="Archivo", menu=archivo_menu) #nombre y dónde agregarla
        archivo_menu.add_command(label='Nuevo')
        archivo_menu.add_command(label="Abrir")
        archivo_menu.add_separator()
        archivo_menu.add_command(label='Salir', command=self.destroy)#para salir utilizar esa función
        
        
        #SUBMENU
        sub_menu = Menu(archivo_menu, tearoff=0) #lo creo, se cambia la raíz
        
        archivo_menu.add_cascade(
                label="Preferencias",
                menu=sub_menu)
        sub_menu.add_command(label='Estilo')
        sub_menu.add_command(label='Color')
 
        # Creamos la segunda categoría
        help_menu = Menu(mimenu, tearoff = False) #Crear categoría
        
        mimenu.add_cascade(label = 'Help', menu = help_menu)
        help_menu.add_command(label = 'About')

        
#Acción:
app = App()
app.mainloop()
```

# 10. NOTEBOOK

El widget Notebook te permite seleccionar páginas de contenido haciendo clic en las pestañas.

Cuando haces clic en una de estas pestañas, el widget Notebook mostrará un panel secundario asociado con la pestaña seleccionada. Normalmente, un panel secundario es un widget Frame.

Para crear un widget Notebook, utiliza la clase ttk.Notebook de la siguiente manera:

```python
class App(tk.Tk): 
    def __init__(self):
        super().__init__()

        # configurando la ventana
        self.title('Mi increíble App OMG') 
        self.geometry('400x300')
        self.resizable(False,False)
        
        #Creo el notebook (aquí creo la opción que existan pestañas)
        self.minotebook = ttk.Notebook(self)
        self.minotebook.pack(pady=10)
        
        
        #lo siguiente es para crear lo que asociaré a las pestañas
        #FRAME(Width, height, padding,borderwidth, relief,): frame['padding'] = (left, top, right, bottom)
        self.frame1 = ttk.Frame(self.minotebook, width=400, height=280)
        self.frame2 = ttk.Frame(self.minotebook, width=400, height=280)
        
        self.frame1.pack()
        self.frame1.pack_propagate(False)
        
        self.frame2.pack()
        self.frame2.pack_propagate(False)
        
        
        # Hago la asociación: añadir frames al notebook

        self.minotebook.add(self.frame1, text='General Information')
        self.minotebook.add(self.frame2, text='Profile')
        

        #LABEL
        self.label1 = ttk.Label(self.frame1, text="Holi", font=("Helvetica",30)) #Creo el Widget y asocio la variable
        self.label1.pack()
        self.boton= ttk.Button(self.frame2, text="presióname",) #Creo el Widget y asocio la variable
        self.boton.pack()


#Acción:
app = App()
app.mainloop()
```

# 11. LabelFrame, es como un frame pero con título

```python
class App(tk.Tk): 
    def __init__(self):
        super().__init__()
        
        
        #Creo la primera cajita llamada Login:
        self.labelframe1=ttk.LabelFrame(self, text="Login:")        
        self.labelframe1.pack()        
        self.login() #Una forma de abreviar y organizar atributos, es volverlos métodos
        
        #Creo la segunda cajuta llamada Operaciones
        self.labelframe2=ttk.LabelFrame(self, text="Operaciones")        
        self.labelframe2.pack()      
        self.operaciones() #Una forma de abreviar y organizar atributos, es volverlos métodos

    def login(self):
        self.label1=ttk.Label(self.labelframe1, text="Nombre de usuario:")
        self.label1.grid(column=0, row=0, padx=4, pady=4)
        self.entry1=ttk.Entry(self.labelframe1)
        self.entry1.grid(column=1, row=0, padx=4, pady=4)
        self.label2=ttk.Label(self.labelframe1, text="Ingrese clave:")        
        self.label2.grid(column=0, row=1, padx=4, pady=4)
        self.entry2=ttk.Entry(self.labelframe1, show="*")
        self.entry2.grid(column=1, row=1, padx=4, pady=4)
        self.boton1=ttk.Button(self.labelframe1, text="Ingresar")
        self.boton1.grid(column=1, row=2, padx=4, pady=4)

    def operaciones(self):
        self.boton2=ttk.Button(self.labelframe2, text="Agregar usuario")
        self.boton2.grid(column=0, row=0, padx=4, pady=4)
        self.boton3=ttk.Button(self.labelframe2, text="Modificar usuario")
        self.boton3.grid(column=1, row=0, padx=4, pady=4)
        self.boton4=ttk.Button(self.labelframe2, text="Borrar usuario")
        self.boton4.grid(column=2, row=0, padx=4, pady=4)

#Acción:
app = App()
app.mainloop()
```

```python
import tkinter as tk
from tkinter import ttk

class App(tk.Tk):
    def __init__(self):
        super().__init__()

        # Crear el widget Scrollbar
        self.scrollbar = tk.Scrollbar(self)
        self.scrollbar.pack(side=tk.RIGHT, fill=tk.Y)

        # Crear el widget Canvas
        self.canvas = tk.Canvas(self, yscrollcommand=self.scrollbar.set)
        self.canvas.pack(side=tk.LEFT, fill=tk.BOTH, expand=True)

        # Configurar el widget Scrollbar para que controle el Canvas
        self.scrollbar.config(command=self.canvas.yview)

        # Crear un Frame dentro del Canvas
        self.frame = ttk.Frame(self.canvas)
        self.canvas.create_window((0, 0), window=self.frame, anchor='nw')

        # Añadir contenido al Frame
        for i in range(50):
            ttk.Label(self.frame, text='Label {}'.format(i)).grid(row=i, column=0)

        # Configurar el tamaño de la ventana y mostrarla
        self.geometry('300x200')
        self.mainloop()

# Acción
app = App()
```
